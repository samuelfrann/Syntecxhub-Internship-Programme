Importing Libraries

```python
import pandas as pd 
import numpy as np
import joblib

from sklearn.pipeline import Pipeline
from sklearn.metrics import r2_score, root_mean_squared_error
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression 
```

Loading Train Set

```python
train = pd.read_csv(r'C:\Users\pc\Documents\JUPYTER ML\Syntecxhub-Internship\House Price Prediction Project\datasets\Busan housing prediction train set.csv')
train.head(5)
```

Feature Selection 

```python
x = train.drop('Housing price', axis=1)
y = train['Housing price']
```

Data Splitting 

```python
x_train, x_validate, y_train, y_validate = train_test_split(x, y, test_size=0.2, random_state=42)
```

Train Set Model Development 

```python
pipe = Pipeline([
    ('scaler', StandardScaler()),
    ('lr', LinearRegression())
])
```

```python
pipe.fit(x_train, y_train)
```

```python
y_pred = pipe.predict(x_validate) 
```

Model Evaluation 

```python
r2_validate = r2_score(y_validate, y_pred)
print("R2 Validation Score:", r2_validate)
```

```python
rmse = root_mean_squared_error(y_validate, y_pred)
print("Validation RMSE:", rmse)
```

```python
coef = pipe['lr'].coef_ 
print("Coefficients:", coef)
```

Loading Test Set 

```python
test = pd.read_csv(r"C:\Users\pc\Documents\JUPYTER ML\Syntecxhub-Internship\House Price Prediction Project\datasets\Busan housing prediction test set.csv")
test.head(5)
```

Model Evaluation on Test Set 

```python
x_test = test.drop('Housing price', axis=1)
y_test = test['Housing price']
```

```python
y_test_pred = pipe.predict(x_test) 
```

```python
r2_test = r2_score(y_test, y_test_pred)
print("R2 Test Score:", r2_test)
```

```python
rmse = root_mean_squared_error(y_test, y_test_pred)
print("Test RMSE:", rmse)
```
